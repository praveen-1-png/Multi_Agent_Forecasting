### Basic array operations
- algebraic operations
- logical operations

```python
a = np.array([[-4,6,8], [9, 3, 1], [1, 8, 5]])
b = np.array([[11,16,18], [90, 37, 81], [61, 48, 75]])
```

```python
# Write a NumPy function to add, subtract, multiply and divide arrays 'a' and 'b'
print("addition of a & b:")
q=a+b
print(q)
print('\n')
print("subtraction of a & b:")
w=a-b
print(w)
print('\n')
print("algebric multiplication of a & b:")
r=np.matmul(a,b)
print(r)
print('\n')
print("division of a & b:")
j=np.linalg.det(b)
if j==0 :
    print("division is not possible")
else:
    v=np.linalg.inv(b)
    print(np.matmul(a,v))
print('\n')


```

```python
# Write a NumPy function to get the sum of the array 'a' along axis=0
print("sum along axis=0:")
print(np.sum(a,axis=0))
print('\n')

# Write a NumPy function to get the sum of the array 'a' along axis=1
print("sum along axis=1:")
print(np.sum(a,axis=1))
print('\n')
```

```python
# Write a NumPy function to reverse the array 'a'
c=np.flip(a)
print("Array after total flip:")
print(c)
print('\n')
# Write a NumPy function to reverse the array 'a' along axis=0
d=np.flip(a,axis=0)
print("Array after flip along axis=0 :")
print(d)
print('\n')
# Write a NumPy function to reverse the array 'a' along axis=1
e=np.flip(a,axis=1)
print("Array after flip along axis=1: ")
print(e)
print('\n')
```

### Indexing and slicing
- NumPy allows indexing array using tuple of nonnegative integers, booleans, array or integers

```python
# Write a NumPy function to index and slice NumPy arrays similar to Python list
# Show all three - indexing by a integer, indexing by a tuple, using range
print("indexing using integer:")
print("the array of a[1]:",a[1])
print('\n')
print("indexing using tuple:")
print("the array of a[(1,2)]:",a[(1,2)])
print('\n')
print("indexing using range:")
print("the array of a[0:2,1:2]:",a[0:2,1:2])
print('\n')
```

```python
# Write a Pandas function to find the mean, median and standard deviation of column 3.
col3_mean = df['petal_length'].mean()
col3_median = df['petal_length'].median()
col3_std = df['petal_length'].std()
print("the required mean:",col3_mean)
print("the required median:",col3_median)
print("the required standard deviation:",col3_std)
print('\n')
```

# Q5 Write a Pandas function to bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
- Less than 3 –> ‘small’
- 3-5 –> ‘medium’
- greater than equal to 5 –> ‘large’

```python

petal_lengths = iris_2d[:, 2]
binned_petals = pd.cut(
    petal_lengths, 
    bins=[0, 3, 5, np.inf], 
    labels=['small', 'medium', 'large'], 
)
print(binned_petals)
print('\n')
```
